```
# Task 04: Convert Table into DataFrame

flyDF1 = spark.sql("SELECT * FROM fly1")
display(flyDF1)
```

```
# Task 05: Create MAP Data type from KV pairs

## Since MAP cannot support KEY = Null, convert all Null's into '0' first before creating MAP data type
flyDF2 = flyDF1.na.fill("0")

flyDF2.show(10, False)

##  Import pyspark library for function 'create_map'
from pyspark.sql.functions import create_map

# Create Map key-value pairs

flyDF3 = flyDF2.select(create_map("k1", "v1", "k2", "v2","k3", "v3","k4", "v4","k5", "v5","k6", "v6","k7", "v7","k8", "v8","k9", "v9","k10", "v10").alias("map_col"))

# Run this command since MAP can't handle Dups without this
spark.sql("set spark.sql.mapKeyDedupPolicy=LAST_WIN")

flyDF3.printSchema()
flyDF3.show(truncate=False)


```

```
 
# Task 06: Create TempView from DataFrame

flyDF3.createOrReplaceTempView("fly4")
```

##### Query:  Display IDENT, ALT and SPEED where ALT > 30000  and LAT is not NULL and IDENT is not NULL and SPEED > 590. Sort by IDENT, SPEED desc

##### Issue is have a lot of Duplicate IDENTs.  Really just need to identify one of them.
##### Note having Duplicate Idents is acceptable answer for Hackathon
##### But if you can have only one Ident with MAX speed your team will get extra credit

```
 %sql
-- Task 08: 
-- Query: Display IDENT, ALT & SPEED where ALT>30000 and LAT not NULL and IDENT not NULL and SPEED>590. Sort by IDENT, SPEED desc


SELECT distinct map_col.ident, map_col.alt, map_col.speed FROM fly4 
  WHERE map_col.alt > 30000 AND map_col.lat IS NOT null
  AND map_col.speed > 590 AND map_col.ident IS NOT null ORDER BY 1,3 DESC
```

```
%sql
select dt.ident,dt.alt, dt.speed from (
select map_col.ident, map_col.alt, map_col.speed, map_col.lat, row_number() over(
partition by map_col.ident order by map_col.speed desc
) as rank from fly4)as dt where dt.rank=1
and dt.alt >30000 and dt.alt is not null
and dt.speed >590 and dt.ident is not null order by dt.ident, dt.speed;
```
